## Importing Data & Libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\BlueMoon\Desktop\Data Projects\Kaggle Survey 2021\Data\kaggle_survey_2021_responses.csv')
```

## Data Cleaning

```python
Q = df.iloc[0,:]
Q
```

```python
df.drop(0,inplace=True)
df.head()
```

```python
for col in df.columns:
    if df[col].str.isnumeric().all():
        df[col] = pd.to_numeric(df[col])
```

```python
df.info()
```

```python
df['Q3'].unique()
```

```python
arab_countries = "Algeria, Bahrain, Comoros, Djibouti, Egypt, Iraq, Jordan, Kuwait, Lebanon, Libya, Mauritania, Morocco, Oman, Palestine, Qatar, Saudi Arabia, Somalia, Sudan, Syria, Tunisia, United Arab Emirates, Yemen"
ar_countries = arab_countries.split(", ")
```

```python
ar_df = df[df['Q3'].isin(ar_countries)]
ar_df.head()
```

## plots of Kaggle Arab Users

###  Q1- What is your age?

```python
ar_age = ar_df['Q1'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10,6))
plt.style.use('bmh')
plt.plot(ar_age.index,ar_age.values)
plt.title('Arab Age Distribution',fontsize=18)
plt.xlabel('Age',fontsize=15)
plt.ylabel('Frequency',fontsize=15);
```

### Q2- What is your gender?

```python
plt.figure(figsize=(8,8))
plt.pie(ar_df['Q2'].value_counts(),labels=ar_df['Q2'].value_counts().index,autopct='%1.1f%%')
plt.title('Gender Distribution',fontsize=18);
```

### Q3- In which country do you currently reside?

```python
ar_country_distribution = ar_df['Q3'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.bar(ar_country_distribution.index,ar_country_distribution.values)
plt.title('Arab country Distribution',fontsize=18)
plt.xlabel('Country',fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Frequency',fontsize=15);
```

### Q4- What is the highest level of formal education?

```python
ar_education_distribution = ar_df['Q4'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.bar(ar_education_distribution.index,ar_education_distribution.values)
plt.title('Highest level of Education Distribution',fontsize=18)
plt.xlabel('Education Level',fontsize=15)
plt.xticks(ar_education_distribution.index,['Bachelor','Master','Doctorate','At college'
                                            ,'I prefer not to answer','high school','Professional doctorate'],rotation=90)
plt.ylabel('Frequency',fontsize=15);
```

### Q5- Select the title most similar to your current role

```python
ar_title_distribution = ar_df['Q5'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.bar(ar_title_distribution.index,ar_title_distribution.values)
plt.title('Title Distribution',fontsize=18)
plt.xlabel('Title',fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Frequency',fontsize=15);
```

### Q6- For how many years have you been writing code and/or programming?

```python
ar_coding_distribution = ar_df['Q6'].value_counts().reset_index()
ar_coding_distribution['i'] = pd.Categorical(ar_coding_distribution['index'], ['< 1 years','1-3 years','3-5 years','5-10 years','10-20 years','20+ years','I have never written code'])
ar_coding_distribution = ar_coding_distribution.sort_values('i')
ar_coding_distribution
```

```python
plt.figure(figsize=(10,6))
plt.plot(ar_coding_distribution['i'],ar_coding_distribution['Q6'])
plt.title('Years of Coding Distribution',fontsize=18)
plt.xlabel('No. of years',fontsize=15)
plt.xticks(rotation=90)
plt.ylabel('Frequency',fontsize=15);
```

### Q7- What programming languages do you use on a regular basis?

```python
Q7=ar_df.columns[ar_df.columns.str.contains('Q7')]
ar_df[Q7]
```

```python
dic_q7 = dict()
for col in Q7:
    key = ar_df[col].value_counts().index[0]
    dic_q7[key] = ar_df[col].value_counts()[0]
dic_q7
```

```python
dic_q7['Other'] = dic_q7['Other'] + dic_q7['Julia'] + dic_q7['None'] + dic_q7['Swift']
q7 = pd.Series(dic_q7).drop(index=['Julia','None','Swift']).sort_values()
q7
```

```python
plt.figure(figsize=(10,10))
plt.pie(q7,labels=q7.index,autopct='%1.1f%%',startangle=90)
plt.title('The programming language used Distribution',fontsize=18);
```
